# Matplotlib
## Introducción

Matplotlib es una biblioteca para producir figuras de calidad de publicación. mpl (para abreviar) se diseñó desde el principio para cumplir dos propósitos:

  1. permitir el control interactivo y multiplataforma de figuras y tramas
  2. Facilite la producción de archivos de gráficos de trama o vector estáticos sin la necesidad de ninguna GUI.
 
Además, mpl, al igual que Python, le da al desarrollador un control completo sobre la apariencia de sus trazados, a la vez que sigue siendo muy útil a través de un poderoso sistema predeterminado.

## Anatomía de una "Plot"

La gente usa "plot" para significar muchas cosas diferentes. Aquí, utilizaremos una terminología consistente (reflejada por los nombres de las clases subyacentes, etc.):

![anatomía basica](images/figure_axes_axis_labeled.png)


La``Figure`` es el contenedor de más alto nivel en la jerarquía. Es la ventana / página general en la que se dibuja todo. Puede tener varias figuras independientes y `` Figure``s puede contener varios `` Axes``.

La mayoría de las tramas ocurren en un `` Axes``. Los ejes son efectivamente el área en la que graficamos los datos y cualquier ticks / labels / etc. asociados con ella. Por lo general, configuraremos un Axes con una llamada a `` subplot`` (que coloca los Axes en una cuadrícula regular), por lo que en la mayoría de los casos, `` Axes`` y `` Subplot`` son sinónimos.

Cada `` Axes`` tiene un `` XAxis`` y un `` YAxis``. Estos contienen los tics, ubicaciones de ticks, etiquetas, etc. En este tutorial, controlaremos principalmente ticks, etiquetas de ticks y límites de datos a través de otros mecanismos, por lo que no tocaremos la parte de `` Axis`` tanto. Sin embargo, vale la pena mencionar aquí para explicar de dónde viene el término `` Axes``.

### Figures

Creemos una figura

````python
fig = plt.figure()
# fig = plt.figure(figsize=plt.figaspect(1.0), facecolor=(1, 0, 0, .1))
ax = fig.add_subplot(111) # We'll explain the "111" later. Basically, 1 row and 1 column.
ax.set(xlim=[0.5, 4.5], ylim=[-2, 8], title='An Example Axes',
       ylabel='Y-Axis', xlabel='X-Axis')
plt.show()
````

# Axes

Todo el trazado se hace con respecto a un [`Axes`](http://matplotlib.org/api/axes_api.html#matplotlib.axes.Axes). Un *Axes* está formado por objetos [`Axis`](http://matplotlib.org/api/axis_api.html#matplotlib.axis.Axis) y muchas otras cosas. Un objeto *Axes* debe pertenecer a *Figure* (y sólo a una *Figure*). La mayoría de los comandos que emitirá serán con respecto a este objeto *Axes*.

Normalmente, configurará una 'Figura', y luego le agregará un 'Axes'.

Puede usar `fig.add_axes`, pero en la mayoría de los casos, encontrará que agregar una subplot se ajustará perfectamente a sus necesidades. (De nuevo, una "subplot" es solo un Axes en un sistema de cuadrícula).

````python
fig = plt.figure()
ax = fig.add_subplot(111) # We'll explain the "111" later. Basically, 1 row and 1 column.
ax.set(xlim=[0.5, 4.5], ylim=[-2, 8], title='An Example Axes',
       ylabel='Y-Axis', xlabel='X-Axis')

plt.show()
````

Observe la llamada a `` set``. Los objetos de Matplotlib suelen tener muchos "setters explícitos", es decir, funciones que comienzan con `` set_ <something> `` y controlan una opción en particular.

Para demostrar esto (y como un ejemplo de la finalización de pestañas de IPython), intente escribir `ax.set_` en una celda de código, luego presione la tecla` <Tab> `. Verás una larga lista de métodos de 'Axes' que comienzan con `set`.

Por ejemplo, podríamos haber escrito la tercera línea arriba como:

````python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlim([0.5, 4.5])
ax.set_ylim([-2, 8])
ax.set_title('A Different Example Axes Title')
ax.set_ylabel('Y-Axis (changed)')
ax.set_xlabel('X-Axis (changed)')
plt.show()
````

Claramente, esto puede volverse rápido. Por lo tanto, el método `set` de Matplotlib puede ser muy útil. Toma cada kwarg que pasas e intenta llamar al "setter" correspondiente. Por ejemplo, `foo.set (bar = 'blah')` llamaría a `foo.set_bar ('blah')`.

Tenga en cuenta que el método `set` no solo se aplica a` Axes`; se aplica a más o menos todos los objetos matplotlib.

Sin embargo, hay casos en los que querrá usar cosas como `ax.set_xlabel ('Some Label', size = 25)` para controlar otras opciones para una función en particular.


## Plotting básico

La mayoría de los diagramas ocurren en un `Axes`. Por lo tanto, si trazas algo en un eje, entonces usarás uno de sus métodos.

Hy varios métodos de trazado. Por ahora, centrémonos en dos métodos: `plot` y` scatter`.

`plot` dibuja puntos con líneas que los conectan. `scatter` dibuja puntos no conectados, opcionalmente escalados o coloreados por variables adicionales.

Como ejemplo básico:

````python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
ax.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], c=[1, 2, 3, 5], marker='^')
ax.set_xlim(0.5, 4.5)
plt.show()
````

## Axes methods vs. pyplot

Curiosamente, casi todos los métodos de un objeto *Axes* existen como una función en el módulo *pyplot* (y viceversa). Por ejemplo, cuando se llama a `plt.xlim (1, 10)`, *pyplot* llama a `ax.set_xlim (1, 10)` en cualquier *Axes* que sea "actual". Aquí hay una versión equivalente del ejemplo anterior usando solo pyplot.

````python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], c=[1, 2, 3, 5], marker='^')
plt.xlim(0.5, 4.5)
plt.show()
````

### Por qué?

“[With pyplot], simple functions are used to add plot elements (lines, images, text, etc.) to the current axes in the current figure.”``


Eso es un poco extraño y tiene menos variales locales, entonces, ¿por qué la mayoría de mis ejemplos no siguen el enfoque de pyplot? Porque [PEP20](http://www.python.org/dev/peps/pep-0020/) "El Zen de Python" dice:

"Explícito es mejor que implícito"

Si bien las tramas muy simples, con guiones cortos se beneficiarían de la concisión del enfoque implícito de las pepitas, al hacer trazados más complicados o trabajar con guiones más grandes, querrá pasar explícitamente el objeto *Axes* y / o *Figure* a operar sobre.

La ventaja de mantener con claridad los ejes con los que trabajamos se hará más obvia cuando comencemos a tener múltiples ejes en una figura.

### Múltiples Axes

Hemos mencionado anteriormente que una figura puede tener más de un 'Axes' en ella. Si desea que sus ejes estén en un sistema de cuadrícula regular, entonces es más fácil usar `plt.subplots (...)` para crear una figura y agregarle los ejes automáticamente.

Por ejemplo:

````python
fig, axes = plt.subplots(nrows=2, ncols=2)
plt.show()
````

`plt.subplots (...)` creó una nueva figura y le agregó 4 subplots. El objeto `axes` que se devolvió es una matriz de objetos numpy 2D. Cada elemento en la matriz es una de las subplots. Están dispuestos como los ves en la figura.

Por lo tanto, cuando queremos trabajar con uno de estos ejes, podemos indexar el conjunto de 'axes' y usar los métodos de ese elemento.

Por ejemplo:

````python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0,0].set(title='Upper Left')
axes[0,1].set(title='Upper Right')
axes[1,0].set(title='Lower Left')
axes[1,1].set(title='Lower Right')

# To iterate over all items in a multidimensional numpy array, use the `flat` attribute
for ax in axes.flat:
    # Remove all xticks and yticks...
    ax.set(xticks=[], yticks=[])
    
plt.show()
````

Una cosa muy buena de `plt.subplots ()` es que cuando se llama sin argumentos, crea una nueva figura con una sola subparcela.

Cada vez que ves algo como

```
fig = plt.figure()
ax = fig.add_subplot (111)
```

Puedes reemplazarlo por:

```
fig, ax = plt.subplots ()
```

Usaremos ese enfoque para el resto de los ejemplos. Es mucho más limpio

Sin embargo, tenga en cuenta que todavía estamos creando una figura y añadiéndole ejes. Cuando comencemos a hacer diseños de parcelas que no pueden describirse mediante 'subparcelas', volveremos a crear la figura primero y luego a agregarle ejes uno a uno.

````python
x = np.linspace(0, 10, 100)
y1, y2, y3 = np.cos(x), np.cos(x + 1), np.cos(x + 2)
names = ['Signal 1', 'Signal 2', 'Signal 3']

fig, axes = plt.subplots(nrows=3)

for ax, y, name in zip(axes, [y1, y2, y3], names):
    ax.plot(x, y, color='black')
    ax.set(xticks=[], yticks=[], title=name)

plt.show()
````
